### statistic describing using Pandas.

Ensured the DataFrame is cleaned and ready to analysis.

```python
df.tail(2)
```

```python
df_cleaned = df
```

```python
# delete string from discounted_price , actual_price 
# change dtype to string first then replace special letter with "" because .str.replce ony work with string dtype.

# column discount_price
df_cleaned["discounted_price"] = df_cleaned["discounted_price"].astype(str)
df_cleaned["discounted_price"] = df["discounted_price"].str.replace("₹", "", regex=False)
df_cleaned["discounted_price"] = df["discounted_price"].str.replace(",", "", regex=False)

# column actual_price
df_cleaned["actual_price"] = df["actual_price"].astype(str)
df_cleaned["actual_price"] = df["actual_price"].str.replace("₹", "", regex=False)
df_cleaned["actual_price"] = df["actual_price"].str.replace(",", "", regex=False)

# column rating_count
df_cleaned["rating_count"] = df_cleaned["rating_count"].astype(str)
df_cleaned["rating_count"] = df_cleaned["rating_count"].str.replace(",", "", regex=False)

# column rating
df_cleaned["rating"] = df_cleaned["rating"].astype(str)
df_cleaned["rating"] = df_cleaned["rating"].str.replace("|", "", regex=False)
df_cleaned["rating"] = df_cleaned["rating"].str.replace("", "0", regex=False)

# # column discount percentage
df_cleaned["discount_percentage"] = df_cleaned["discount_percentage"].astype(str)
df_cleaned["discount_percentage"] = df_cleaned["discount_percentage"].str.replace("%", "", regex=False)
```

```python
# convert dtype to appropriate one.
# 2 ways to convert multiple column dtype.

# 1) 1 line solution
# df_cleaned[["discounted_price", "actual_price"]] = df_cleaned[["discounted_price", "actual_price"]].astype(str)

# 2) Dictionary solution
# df_cleaned = df_cleaned.astype({
#     "discounted_price" : str,
#     "actual_price" : str
# })

# so I use dictionary solution because it easy to read
df_cleaned = df_cleaned.astype({
    "discounted_price" : float,
    "actual_price" : float,
    "rating_count" : float,
    "rating" : float,
    "discount_percentage" : float
})


```

```python
# divide discount_percentage with 100 to make it a portion of percentage.
df_cleaned["discount_percentage"] = df_cleaned["discount_percentage"]/100
```

```python
df_cleaned.head(3)
```

```python
# Add column "rating_group" to DataFrame
condition = [
    df_cleaned["rating"] > 40,
    (df_cleaned["rating"] >= 25) & (df_cleaned["rating"] <= 40),
    df_cleaned["rating"] <25 
]

text = ["high", "middle", "low"]
df_cleaned["rating_rank"] = np.select(condition, text, default="unknown")
```

```python
# plt histogram function
def plt_hist_fn(
        x,
        bins= 3,
        width= 0.5,
        color = "orange",
        edgecolor = "black",
        alpha = 0.7,
        title= "no title",
        xlabel= "no xlabel",
        ylabel= "no ylabel",
        grid= False
    ):  
    counts, bins, patches = plt.hist(
        x,
        bins=bins,
        width=width,
        color=color,
        edgecolor=edgecolor,
        alpha=alpha
    )
    
    for plt_hist_x_fn in patches:
        plt_hist_height_fn = plt_hist_x_fn.get_height()
        plt.text(
            plt_hist_x_fn.get_x() + plt_hist_x_fn.get_width()/2,
            plt_hist_height_fn,
            str(int(plt_hist_height_fn)),
            va= "bottom",
            ha= "center"
        )
        
    plt.grid(grid)
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.show()
```

```python
# call plt_hist_fn
plt_hist_fn(
    x = df_cleaned[["rating_rank"]],
    title= "my plt hist",
    xlabel= "rank",
    ylabel= "amount"
    )
```

```python
# Create bar plot function 
# using data from --> plt_bar_info

def plt_bar_fn(
        x,
        height,
        color= "#C79FEF",
        alpha= 0.7,
        edgecolor= "black",
        title= "Untitled",
        xlabel= None,
        ylabel= None,
        grid= False,
        round_num=2,
        fontweight= None
):
    plt_bar_chart_fn = plt.bar(
        x,
        height,
        color=color,
        alpha=alpha,
        edgecolor=edgecolor
    )
    for plt_bar_x_fn in plt_bar_chart_fn:
        plt_bar_height_fn = plt_bar_x_fn.get_height()
        plt.text(
            plt_bar_x_fn.get_x() + plt_bar_x_fn.get_width()/2,
            plt_bar_height_fn,
            str(round(plt_bar_height_fn, round_num)),
            va= "bottom",
            ha= "center"
        )
    plt.grid(grid)
    plt.title(title, fontweight=fontweight)
    plt.xlabel(xlabel, fontweight=fontweight)
    plt.ylabel(ylabel, fontweight=fontweight)
    plt.show()
```

```python
# calling plt bar function
# using data from --> plt_bar_info
plt_bar_fn(x= plt_bar_info["rating_rank"], height= plt_bar_info["average_discount%"], title= "My bar chart", xlabel= "rank", round_num= 3, fontweight="bold")
```

```python
### Create bar plot with ax method which give less manual and get cleaner code.
# Step 1 : subplot
fig, ax = plt.subplots()

# Step 2 : Create bar chart
my_bars = ax.bar(
    x= plt_bar_info["rating_rank"],
    height= plt_bar_info["average_discount%"],
    color="#C79FEF",
    alpha=0.7,
    edgecolor="black"
)

# Step 3 : Add label automatically
ax.bar_label(my_bars, fmt="%.2f") # .2 represent the 

# Step 4 : Customize chart
ax.set_title("Average %discount per rating rank", y=1.1, fontweight= "bold")
ax.set_xlabel("rating rank", fontweight= "bold")
ax.set_ylabel("%discount", fontweight= "bold")
ax.grid(False)

# Step 5 : Deploy chart
plt.show()
```

```python
# Create bar plot function using ax method

def ax_bar_fn(
    x,
    height,
    color= "#C79FEF",
    alpha= 0.7,
    edgecolor= "black",
    round_num= 2,
    title= "Untitled",
    xlabel= None,
    ylabel= None,
    fontbold= None,
    grid= False
):
    # subplots
    fig, ax = plt.subplots()
    
    # create bar chart
    my_bars_fn = ax.bar(
        x,
        height,
        color=color,
        alpha=alpha,
        edgecolor=edgecolor
    )
    
    # data labels
    ax.bar_label(my_bars_fn, fmt=f"%0.{round_num}f")
    
    # customize chart
    ax.set_title(title, y=1.1, fontweight=fontbold)
    ax.set_xlabel(xlabel, fontweight=fontbold)
    ax.set_ylabel(ylabel, fontweight=fontbold)
    ax.grid(grid)

    # Deploy
    plt.show()
```

```python
# Calling ax bar function
# using data from --> plt_bar_info
ax_bar_fn(x= plt_bar_info["rating_rank"], height= plt_bar_info["average_discount%"], round_num=3, title= "My ax bar", fontbold="bold")
```

```python
# create line plot function.
def ax_plot_fn(
        x,
        y,
        color= "red",
        alpha= 0.6,
        size= (8,6),
        title= "Untitled",
        xlabel= None,
        ylabel= None,
        fontbold= None,
        grid= False
):
    # subplots.
    fig, ax = plt.subplots(figsize=size)

    # create line plot.
    line_plot_fn = ax.plot(
        x,
        y,
        color=color,
        alpha=alpha
    )

    # customize line plot.
    ax.set_title(title, y=1.1, fontweight=fontbold)
    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)
    ax.grid(grid)

    # deploy
    plt.show()
    
```
