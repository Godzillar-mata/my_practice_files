### statistic describing using Pandas.

##### Clean data.

```python
# change all column to proper column name.
df_renamed = df.rename(columns={
    "Company Names" : "company_name",
    "Cars Names" : "cars_name",
    "Engines" : "engines",
    "CC/Battery Capacity" : "cc",        # CC.
    "HorsePower" : "horsepower",                    # hp.
    "Total Speed" : "total_speed",                  # km/h.
    "Performance(0 - 100 )KM/H" : "performance",    # sec.
    "Cars Prices" : "cars_prices",                  # dollars.
    "Fuel Types" : "fuel_types",
    "Seats" : "seats",                              # seat.
    "Torque" : "torque"                             # Nm.
})
```

```python
df_cleaned = df_renamed
```

```python
### 1) column : battery_capacity. 

# make sure the column is string type.
df_cleaned["cc"] = df_cleaned["cc"].astype(str)

# extract only digits.
df_cleaned["cc"] = df_cleaned["cc"].str.extract(r"(\d+)")

# drop null value and change dtype to integer.
df_cleaned["cc"] = df_cleaned["cc"].dropna().astype(int)
```

```python
### 2) column : horsepower.

# make sure the column is string type.
df_cleaned["horsepower"] = df_cleaned["horsepower"].astype(str)

# delete text hp and fill average value for range data.
# function to clean hp column.
def clean_hp(value):
    import re
    # find all number in str.
    numbers = re.findall(r"\d+", str(value)) # r"\d+" only grab degits 0-9 not include decimal.
    numbers = [int(n) for n in numbers]
    if len(numbers) == 0:
        return None
    elif len(numbers) == 1:
        return numbers[0]
    else:
        return sum(numbers) // len(numbers)
    
# Apply function.
df_cleaned["horsepower"] = df_cleaned["horsepower"].apply(clean_hp)

# change dtype to integer.
df_cleaned["horsepower"] = df_cleaned["horsepower"].astype(int)
```

```python
### 3) column total_speed

# make sure column is string type.
df_cleaned["total_speed"] = df_cleaned["total_speed"].astype(str)

# delete km/h
df_cleaned["total_speed"] = df_cleaned["total_speed"].str.extract(r"(\d+)")

# convert to int
df_cleaned["total_speed"] = df_cleaned["total_speed"].astype(int)
```

```python
### 4) column performance.

# make sure the column is string type.
df_cleaned["performance"] = df_cleaned["performance"].astype(str)

# function to clean : detect number and calculate average for a range number.
def clean_performance(value):
    import re
    # Find all number in string.
    numbers = re.findall(r"\d+(?:\.\d+)", str(value)) # r"\d+(?:\.\d+)" will grab include decimal.  
    numbers = [float(n) for n in numbers]
    if len(numbers) == 0:
        return None
    elif len(numbers) == 1:
        return numbers[0]
    else:
        return sum(numbers) // len(numbers)

# apply function.
df_cleaned["performance"] = df_cleaned["performance"].apply(clean_performance)

# drop na then convert dtype to int.
df_cleaned["performance"] = df_cleaned["performance"].dropna().astype("Float64")
```

```python
### 5) column : cars_prices
# 

# In column cars_prices have different unit price so we need to add 1. column currency, 2. column currency multiplier , 3. cars_prices_dollars

# add column currency
df_cleaned["currency"] = df_cleaned["cars_prices"].astype(str).str[0]

# add column currency multiplier
df_cleaned["currency_mulp"] = df_cleaned["currency"].map({"$" : 1, "€" : 1.169, "N" : 0})

# show cars_prices as int/float.
def clean_cars_prices(value):
    import re
    # find number with comma and decimals.
    numbers = re.findall(r"[\d,]+(?:\.\d+)?", str(value))
    if not numbers:
        return pd.NA
    # remove comma and convert to float
    vals = [float(n.replace(",", "")) for n in numbers]
    avg = sum(vals)/len(vals)
    return int(round(avg, 0))


df_cleaned["cars_prices"] = df_cleaned["cars_prices"].apply(clean_cars_prices).astype("Int64")

# add column new_cars_prices which convert all cars prices in the same currency(dollars).
df_cleaned["new_cars_prices"] = df_cleaned["cars_prices"] * df_cleaned["currency_mulp"]
```

```python
# replace 2+2 with 4 in str.
df_cleaned["seats"] = df_cleaned["seats"].str.replace("2+2", "4")

# clean range seat like 2-12, 2-6 and replace it with maximum seat.
df_cleaned["seats"] = df_cleaned["seats"].apply(lambda x: x.split("–")[-1] if "–" in x else x)

# change column seats to int.
df_cleaned["seats"] = df_cleaned["seats"].astype(int)
```

```python
# replace Nm with space.
df_cleaned["torque"] = df_cleaned["torque"].str.replace("Nm", "")

# replace range value with mean value.
def clean_torque(value):
    import re
    # find number include decimal.
    nums = re.findall(r"[\d,]+(?:\.\d+)?", str(value))
    nums = [float(n.replace(",", "")) for n in nums]
    if len(nums) == 0:
        return None
    elif len(nums) == 1:
        return int(nums[0])
    else:
        return int(sum(nums) / len(nums))
# apply function.
df_cleaned["torque"] = df_cleaned["torque"].apply(clean_torque)

# drop missing value and chage type.
df_cleaned["torque"] = df_cleaned["torque"].dropna().astype(int)
```
