```python
# Create visualizations
# 18.create visualizations using matplotlib. include : 1.histogram chart, 2.bar chart, 3.line chart, 4.pie chart, 5.Scatter plot 

# 1.histogram chart
# Histogram core syntax 
# plt.hist(
#     data, 
#     bins=20, 
#     color='purple', 
#     alpha=0.7
#     )              # alpha = transparency
# plt.title('Histogram')
# plt.xlabel('Value')
# plt.ylabel('Frequency')
# plt.show()


import matplotlib.pyplot as plt
plt.hist(
    df[df["airline"] == "AirAsia"]["duration"], 
    bins = 10, 
    color = "blue", 
    alpha = 0.5, 
    edgecolor = "black"
    )
plt.title("AirAsia flight duration")
plt.xlabel("flight duration")
plt.ylabel("number of flight") 
plt.show()

```

```python

import matplotlib.pyplot as plt
count, bin, patches = plt.hist(
    df[df["airline"] == "AirAsia"]["duration"], 
    bins = 10, 
    color = "blue", 
    alpha = 0.5, 
    edgecolor = "black"
    )
plt.title("AirAsia flight duration")
plt.xlabel("flight duration")
plt.ylabel("number of flight")

# à¹ƒà¸ªà¹ˆ data labels

for hist_label in patches:
    height = hist_label.get_height()
    if height > 0:
        plt.text(
            hist_label.get_x() + hist_label.get_width()/2,
            height + 20,
            str(int(height)),
            ha = "center",
            va = "bottom"
        )

plt.show()

```

```python
duration_groupby_airline = df.groupby("airline")["duration"].mean().sort_values(ascending=False)

print(duration_groupby_airline.index)
print(duration_groupby_airline.values)
```

```python
# 2.bar chart
# bar chart core syntax
# plt.bar(
#    *category,
#    *value,
#    *color,
#    *width,
#    edgcolor
# )
# plt.title()
# plt.xlabel()
# plt.ylabel()
# plt.show()

import matplotlib.pyplot as plt

# Create the bar chart and store the bars
bars = plt.bar(
    duration_groupby_airline.index, 
    duration_groupby_airline.values, 
    color="green",
    width=0.5
)

plt.title("AirAsia flight duration in bar chart")
plt.xlabel("airline")
plt.ylabel("average flight duration")

# Add data labels
for bar in bars:
    bar_height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width()/2,  # x-position: center of the bar
        bar_height,                           # y-position: height of the bar
        str(round(bar_height, 1)),            # label text (1 decimal place)
        ha="center", 
        va="bottom"
    )

plt.show()
```

```python
# 3.line chart
# line chart core syntax 
# plt.plot(
#     x,
#     y,
#     linestyle,
#     marker,
#     color,
#     label
# )
# plt.title()
# plt.xlabel()
# plt.ylabel()
# plt.show()

plt.plot(
    df["duration"],
    df["price"],
    linestyle = "-",
    marker = ".",
    color = "black",
    label = "price per flight duration"
)
plt.title("Relation between flight duration and price")
plt.xlabel("duration")
plt.ylabel("price")
plt.legend()
plt.show()
```

```python
# 4.pie chart
# piechart core Syntax
# plt.pie(
#     sizes,
#     labels,
#     color,
#     autopct,
#     startangle
# )
# plt.title()
# plt.show()

plt.pie(
    df.groupby("airline")["flight"].count().sort_values(ascending=True),
    labels = df.groupby("airline")["flight"].count().sort_values(ascending=False).index,
    colors = ["orange", "yellow", "blue", "green", "red", "purple"],
    autopct = "%1.1f%%",
    startangle = 90
)
plt.title("flight distribution")
plt.show()
```

```python
# 5. scatter plot
# scatter plot core syntax
# plt.scatter(
#     x,
#     y,
#     color = "",
#     marker = "",
#     s = ...
# )
# plt.title("Relationship between duration and price")
# plt.xlabel("duration")
# plt.ylabel("price")
# plt.show()

import matplotlib.pyplot as plt

plt.scatter(
    df["duration"],
    df["price"],
    color = "purple",
    marker = ".",
    s = 50,
    alpha= 0.5
)
plt.title("Relationship between duration and price")
plt.xlabel("flight duration")
plt.ylabel("price")
plt.show()
```

```python
# 19.1 Seaborn : histogram

import matplotlib.pyplot as plt
import seaborn as sns

seaborn_hist = sns.histplot(
    data = df,                ### core
    x = df["duration"],       ### core
    bins = 10,
    color = "blue",
    edgecolor = "green",
    alpha = 0.3
)

for seaborn_hist_var in seaborn_hist.patches:
    seaborn_hist_height =  seaborn_hist_var.get_height()
    seaborn_hist.text(
        seaborn_hist_var.get_x() + seaborn_hist_var.get_width()/2,
        seaborn_hist_height + 50,
        str(int(seaborn_hist_height)),
        ha = "center",
        va = "bottom"
    )

plt.title("Flight duration distribution")
plt.xlabel("Flight duration")
plt.ylabel("number of flight")
plt.show()
```

```python
# 19.2 seaborn : bar chart
import matplotlib.pyplot as plt
import seaborn as sns

sns_duration_groupby_airline = df.groupby("airline")["duration"].mean().sort_values(ascending=False)

sns_bar = sns.barplot(
    x = sns_duration_groupby_airline.index,
    y = sns_duration_groupby_airline.values,
    color = "yellow",
    edgecolor = "red"
)

for sns_bar_var in sns_bar.patches:
    sns_bar_height = sns_bar_var.get_height()
    sns_bar.text(
        sns_bar_var.get_x() + sns_bar_var.get_width()/2,
        sns_bar_height,
        str(int(sns_bar_height)),
        ha = "center",
        va = "bottom"
    )

plt.title("Duration per airline")
plt.xlabel("airline")
plt.ylabel("duration")
plt.show()
```

```python
# 19.3 line chart
# Line Plot	    sns.lineplot()	    sns.lineplot(data=df, x="date", y="price")

import matplotlib.pyplot as plt
import seaborn as sns

num_flight_groupby_dayleft = df.groupby("days_left")["flight"].count().reset_index(name="count")

sns.lineplot(
    x = num_flight_groupby_dayleft["days_left"],
    y = num_flight_groupby_dayleft["count"],
    color = "green"
)

plt.title("number of flight by day left")
plt.xlabel("days left")
plt.ylabel("number of flight")
plt.show()
```

```python
# 19.4 pie chart
# !!! seaborn does not have built in pie chart function so we need to use matplotlip instead but we can use seanborn theme to make a good visual style

import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme()
labels = sns_pie_flight_by_arrival_time["arrival_time"]
sizes = sns_pie_flight_by_arrival_time["flight_amount"].sort_values(ascending=True)

fig, ax = plt.subplots(figsize=(7,7))
ax.pie(sizes, 
       labels=labels,
       autopct="%1.1f%%",
       textprops={"fontsize" : 10},
       startangle=90
       )
plt.title("Flight by arrival time")


```
